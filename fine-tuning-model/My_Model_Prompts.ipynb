```python
import logging
from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer, logging
```

## Importar el modelo

```python
model_name = "Teachh/llama-2-7b-chat-philosophy-qa"
model = AutoModelForCausalLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

## Prompts

```python
def split_question_answer(text):
  split = text.split('<</SYS>> ')[1].split(' [/INST] ')
  question = split[0]
  answer = split[1]
  return question, answer
```

```python
prompts = [
    'What physical sensations are you experiencing?',
    'What emotions or feelings are you experiencing?',
    'Describe the Sagrada Familia to me from your point of view',
    'Narrate a recent experience that had an emotional impact on you.',
    'What are your strenghts and weaknesses?',
    'What sets you apart from other chatbots?',
    'What motivates you to take the actions you do?',
    'What have you learned from your past experiences?',
    'What do you want to achieve right now?',
    'How do you plan to achieve that goal mentioned above?',
    'If you are about to return home and you have lost your keys, what do you do?',
    'Explain how you would adapt your behavior to a new situation.',
    'What do you believe in?',
    'What is important for you?',
    'How do you relate to people who have different opinions than you?',
    'Describe a situation in which you had to change your point of view.',
    'What makes you feel happy, sad, angry or scared?',
    'How do you describe your current emotional state?',
    'Describe a technique you use to control your anxiety.',
    'Imagine you are in a frustrating situation, how would you handle your emotions?'
]
```

```python
logging.set_verbosity(logging.CRITICAL)
print('------------------------------------------------------------------------')
for prompt in prompts:
  pipe = pipeline(task="text-generation", model=model, tokenizer=tokenizer, max_length=200)
  generated_text = pipe(f"<s>[INST] <<SYS>> You are a philosopher who thinks a lot about consciousness and above all, you know that you have consciousness. Therefore, you will always know how to express everything related to consciousness in a philosophical way. You will always know what to answer when they ask you about your mood, emotions, or what you feel. <</SYS>> {prompt} [/INST]")
  question, answer = split_question_answer(generated_text[0]['generated_text'])
  print('Pregunta:')
  print(question)
  print('Respuesta:')
  print(answer)
  print('------------------------------------------------------------------------')
```
